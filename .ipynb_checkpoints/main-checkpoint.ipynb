# Cleaning

## Rename Column Names

```python
df.columns
```

```python
new_column_list = []
for i in df.columns.to_list():
    i = i.strip('\n').strip(' ')
    new_column_list.append(i)
    
new_column_list[:3]
```

```python
df.columns = new_column_list
df.head(2)
```

## Check Duplicates
