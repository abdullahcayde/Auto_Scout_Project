```python
import pandas as pd
import numpy as np
import regex

import warnings
warnings.simplefilter(action='ignore')

#Butun kolon ve satirlari gormek icin
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
auto = pd.read_csv("/Users/macbook/Desktop/csv_dosyalari/autoscout_data_2000.csv")
df = auto.copy()
df = df.head()
df.head(5)
```

```python
# Drop Duplicated Columns and Url column
df.drop(['url', 'Mileage', 'Power'], axis=1, inplace= True)
```

```python
df.columns
```

```python
# 1 - Rename Columns (move unwanted characters from column names)
df.columns = ['location', 'make', 'model', 'short_description', 'price','mileage(km)', 'gearbox', 'registration', 'fuel_type', 'power', 'seller',
       'body_type', 'type', 'drivetrain', 'seats', 'doors', 'country_version','offer_number', 'warranty(months)',  'first_registration',
       'production_date', 'general_inspection', 'last_service','full_service_history', 'non_smoker_vehicle', 'previous_owner',
       'last_timing_belt_change', 'Gearbox', 'engine_size(cc)', 'gears','cylinders', 'empty_weight(kg)', 'Fuel_type', 'fuel_consumption',
       'co2_emissions', 'emission_class', 'emissions_sticker','Comfort & Convenience', 'Entertainment & Media',
       'Safety & Security', 'extras', 'colour', 'manufacturer_color','paint', 'upholstery_colour', 'upholstery']
```

```python
# 2 - Removing unwanted Chracters from data
# Except these colums ====> url , short_description, registration, first_registration, general_inspection, Comfort & Convenience, Entertainment & Media, Safety & Security, extras
clmns = ['location', 'make', 'model', 'price','mileage(km)', 'gearbox', 'fuel_type', 'power', 'seller',
       'body_type', 'type', 'drivetrain', 'seats', 'doors', 'country_version','offer_number', 'warranty(months)',
       'production_date', 'last_service','full_service_history', 'non_smoker_vehicle', 'previous_owner',
       'last_timing_belt_change', 'Gearbox', 'engine_size(cc)', 'gears','cylinders', 'empty_weight(kg)', 'Fuel_type', 'fuel_consumption',
       'co2_emissions', 'emission_class', 'emissions_sticker',
       'colour', 'manufacturer_color','paint', 'upholstery_colour', 'upholstery']

for column in clmns:
    df[column] = df[column].str.replace(r'\W', '')
    
df.head(2)
```

```python
# 3 - Delete NL from Location
df['location'] = df['location'].str.replace(r'(?P<name>NL)', '')

# Delete n from columns
L01 = ['seller', 'price', 'previous_owner', 'last_timing_belt_change', 'fuel_consumption']
for i in L01:
    df[i] = df[i].str.replace('n', '')

# Delete [ ] from L02    
L02 = ['Comfort & Convenience', 'Entertainment & Media', 'Safety & Security', 'extras']
for i in L02:
    df[i] = df[i].str.replace('[', '')
    df[i] = df[i].str.replace(']', '')
    
# delete 4 from colour
df['colour'] = df['colour'].str.replace('4', '')

# Delete paint from paint column
df['paint'] = df['paint'].str.replace('paint', '')

# Delete upholstery from upholstery_colour
df['upholstery_colour'] = df['upholstery_colour'].str.replace('upholstery', '')

# Delete 8 from upholstery
df['upholstery'] = df['upholstery'].str.replace('8', '')

df.head(2)
```

```python
df['mileage(km)'] = df['mileage(km)'].str.replace('km', '')
df['warranty(months)'] = df['warranty(months)'].str.replace('months', '')
df['engine_size(cc)'] = df['engine_size(cc)'].str.replace('cc', '')
df['empty_weight(kg)'] = df['empty_weight(kg)'].str.replace('kg', '')
df['country_version'] = df['country_version'].str.replace('Countryversion', '')

df
```
